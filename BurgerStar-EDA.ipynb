```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
data=pd.read_csv('Static-Dataset-Transaction-Burger-Star-(Outlet A) .csv')
```

```python
data.head()
```

```python
data.describe()
```

```python
burger=pd.DataFrame(data[['Amount Burger']].groupby(data['Type Burger']).sum().sort_values(by='Amount Burger',ascending=False).reset_index())
sns.barplot(x=burger['Type Burger'],y=burger['Amount Burger'])
```

```python
sns.countplot(data['Type Burger'])
```

```python
drink=pd.DataFrame(data[['Amount Drink']].groupby(data['Type Drink']).sum().sort_values(by='Amount Drink',ascending=False).reset_index())
sns.barplot(x=drink['Type Drink'],y=drink['Amount Drink'])
```

```python
sns.countplot(data['Payment'].sort_values())
```

```python
sns.countplot(data['Gender'].sort_values())
```

```python
bur=pd.DataFrame(data[['Amount Burger']].groupby([data['Type Burger'],data['Gender']]).sum().sort_values(by='Amount Burger',ascending=False).reset_index())
sns.barplot(x=bur['Type Burger'],y=bur['Amount Burger'],hue=bur['Gender'])

```

```python
dri=pd.DataFrame(data[['Amount Drink']].groupby([data['Type Drink'],data['Gender']]).sum().sort_values(by='Amount Drink',ascending=False).reset_index())
sns.barplot(x=dri['Type Drink'],y=dri['Amount Drink'],hue=dri['Gender'])
```

```python
sns.countplot(data['Payment'].sort_values(),hue=data['Gender'])
```
